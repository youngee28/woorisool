```python
import pandas as pd
```

```python
csv_file = pd.read_csv('new_tb_info.csv', encoding='cp949')
csv_file
```

```python
query_list = []

for i in range(len(csv_file)):
  items = csv_file.iloc[i].to_list() # [product_id, type_id, keyword]
  query = "UPDATE tb_info SET INFO_KEYWORD = '{}' WHERE PRODUCT_ID = '{}' AND TYPE_ID = '{}'".format(items[2], items[0], items[1])
  query_list.append(query)

query_list = ';\n'.join(query_list)

file = open('update.sql', 'w')    # hello.txt 파일을 쓰기 모드(w)로 열기. 파일 객체 반환
file.write(query_list)      # 파일에 문자열 저장
file.close()
```
